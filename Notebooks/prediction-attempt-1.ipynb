```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
#reading csv file from db
df = pd.read_csv('../input/boston-crime-data/crime.csv', encoding='latin-1')
df.head()
```

# Feature Engineering

```python
df.info()
```

```python
#filling null values in shooting feature as '0' and replacing 'Y' values with '1'
df['SHOOTING']= df['SHOOTING'].fillna(value='0')
df['SHOOTING'] = df['SHOOTING'].replace('Y',1)
df
```

```python
#droping Nan values from Lat and Long 
#droping from Lat and Long with the values of '-1' --removing outliers
# it reduced the number of rows from 327.727 to 306.320
df = df.dropna(subset = ['Lat', 'Long'])
df = df[df['Lat'] > 0]
df
```

```python
df.describe()
```

```python
df.info()
```

```python
#creating date_time format and drop occured_on_date column
df["OCCURRED_ON_DATE"]=pd.to_datetime(df["OCCURRED_ON_DATE"],infer_datetime_format=True)
df["OCCURRED_DAY"]=df["OCCURRED_ON_DATE"].apply(lambda x: x.date())
df = df.drop(['OCCURRED_ON_DATE'],axis = 1)
df.info()
```

```python
#for model 1 creating a dataframe which name is df_m_1, sorting the dates and drop na rows from DISTRICT feature
df_m_1 = df.copy()
df_m_1 = df_m_1.sort_values(by =['OCCURRED_DAY'], ascending=False)
df_m_1['OCCURRED_DAY']= pd.to_datetime(df_m_1['OCCURRED_DAY'])
df_m_1 = df_m_1.dropna(subset=['DISTRICT'])
df_m_1
```

```python
df_m_1.info()
```

```python
df['DAY_OF_WEEK'].value_counts()
```

```python
# day_of_week label encoding

df_m_1['DAY_OF_WEEK'] = df_m_1['DAY_OF_WEEK'].map({
    'Tuesday':4, 
    'Saturday':2, 
    'Monday':3, 
    'Sunday':1, 
    'Thursday':5, 
    'Wednesday':6,
    'Friday':7
})
df_m_1
```

```python
# df_m_1['UCR_PART'] = df_m_1['UCR_PART'].map({
#     'Part One':3, 
#     'Part Two':2, 
#     'Part Three':1, 
#     'Other':0, 

# })
# df_m_1
```

```python
# creating summer, HOUR_12_16_17_18_19, HOUR_3_4_5_6(least hours), RIVER | CENTRE (most occured streets), RA (most occured reporting areas) features which have the highest crime occured

df_m_1['Summer']=0
df_m_1['Summer']=df_m_1["MONTH"].apply(lambda x: 1 if x==6 or x==7 or x==8 or x==9 else 0)


df_m_1['HOUR_12_16_17_18_19']=0
df_m_1['HOUR_12_16_17_18_19']=df_m_1["HOUR"].apply(lambda x: 1 if x==12 or x==16 or x==17 or x==18 or x==19 else 0)
df_m_1

df_m_1['HOUR_3_4_5_6']=0
df_m_1['HOUR_3_4_5_6']=df_m_1["HOUR"].apply(lambda x: 1 if x==3 or x==4 or x==5 or x==6 else 0)
df_m_1

df_m_1['RIVER | CENTRE'] = 0
df_m_1['RIVER | CENTRE'] = df_m_1["STREET"].apply(lambda x: 0 if x=='RIVER ST' or x=='CENTRE ST' else 1)
df_m_1

df_m_1['RA'] = 0
df_m_1['RA'] = df_m_1["REPORTING_AREA"].apply(lambda x: 1 if x=='143' or x=='134' or x=='186' or x=='111' else 1)
df_m_1
```

```python
#one-hot encoding for district and UCR_Part columns
def get_dummies(dataframe):
    
    #for DISTRICT column
    
#     DISTRICT = pd.get_dummies(dataframe.DISTRICT, prefix = 'DIS')
#     dataframe = pd.concat([dataframe,DISTRICT],axis = 1)
#     dataframe = dataframe.drop(['DISTRICT'],axis = 1)
    
    #for UCR PART column
    
    UCR_PART = pd.get_dummies(dataframe.UCR_PART, prefix = 'UCR')
    dataframe = pd.concat([dataframe,UCR_PART],axis = 1)
    dataframe = dataframe.drop(['UCR_PART'],axis = 1)
    
    return dataframe
df_m_1 =get_dummies(df_m_1)
df_m_1
```

```python
# defining function for getting the classification metrics easily
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

def classification_metrics(y_test, predict):
    print("Confusion Matrix:\n")
    print(confusion_matrix(y_test, predict))
    print("\nAccuracy: ", accuracy_score(y_test, predict))
    print("\nClassification report:\n")
    print(classification_report(y_test, predict))
```
